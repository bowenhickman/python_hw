```python
import json

# define function meanReversionStrategy
def meanReversionStrategy(prices):
    window = 5
    in_position = False
    total_profit = 0.0
    first_buy_price = None
    current_buy_price = None

    # mean reversion average trading strategy logic
    for i in range(window-1, len(prices)):
        avg = sum(prices[i-window+1:i+1])/window
        avg = round(avg, 2)
        price = prices[i]

        # Strategy logic
        if price < avg * 0.98:
            if not in_position:
                current_buy_price = price
                print(f"buy at: {price:.2f}")
                in_position = True
                if first_buy_price is None:
                    first_buy_price = price
        elif price > avg * 1.02:
            if in_position:
                profit = price - current_buy_price
                total_profit += profit
                print(f"sell at: {price:.2f}")
                print(f"trade profit: {profit:.2f}")
                in_position = False

    # Calculate returns
    if first_buy_price and first_buy_price != 0:
        percent_return = (total_profit / first_buy_price) * 100
    else:
        percent_return = 0.0

    # Final output
    print("-----------------------")
    print(f"Total profit: {round(total_profit, 2):.2f}")
    print(f"First buy: {first_buy_price:.2f}" if first_buy_price else "First buy: None")
    print(f"Percent return: {round(percent_return, 2):.2f}%")

    return round(total_profit, 2), round(percent_return, 2)

# define function simpleMovingAverageStrategy
def simpleMovingAverageStrategy(prices):
    window = 5
    in_position = False
    total_profit = 0.0
    first_buy_price = None
    current_buy_price = None

    # simple moving average trading strategy logic
    for i in range(window-1, len(prices)):
        avg = sum(prices[i-window+1:i+1])/window
        avg = round(avg, 2)
        price = prices[i]

        # Strategy logic
        if price > avg:
            if not in_position:
                current_buy_price = price
                print(f"buy at: {price:.2f}")
                in_position = True
                if first_buy_price is None:
                    first_buy_price = price
        elif price < avg:
            if in_position:
                profit = price - current_buy_price
                total_profit += profit
                print(f"sell at: {price:.2f}")
                print(f"trade profit: {profit:.2f}")
                in_position = False

    # Calculate returns
    if first_buy_price and first_buy_price != 0:
        percent_return = (total_profit / first_buy_price) * 100
    else:
        percent_return = 0.0

    # Final output
    print("-----------------------")
    print(f"Total profit: {round(total_profit, 2):.2f}")
    print(f"First buy: {first_buy_price:.2f}" if first_buy_price else "First buy: None")
    print(f"Percent return: {round(percent_return, 2):.2f}%")

    return round(total_profit, 2), round(percent_return, 2)

# define function saveResults
def saveResults(data):
    with open('results.json', 'w') as f:
        json.dump(data, f, indent=4)

# create list to store 10 tickers
tickers = ['AAPL', 'GOOG', 'ADBE', 'TSLA', 'BA', 'CMCSA', 'CSCO', 'CVX', 'JPM', 'AMZN']

# create dictionary called results to store prices, profits and return percentages
results = {}

# loop through the list of tickers
for ticker in tickers:
    try:
        # load prices from a file <ticker>.txt
        with open(f"/content/{ticker}.txt", 'r') as f:
            prices = [round(float(line.strip()), 2) for line in f.readlines()]

        # store them in the results dictionary with the key "<ticker>_prices"
        results[f"{ticker}_prices"] = prices

        # call meanReversionStrategy
        mr_profit, mr_returns = meanReversionStrategy(prices)
        results[f"{ticker}_mr_profit"] = mr_profit
        results[f"{ticker}_mr_returns"] = mr_returns

        # call simpleMovingAverageStrategy
        sma_profit, sma_returns = simpleMovingAverageStrategy(prices)
        results[f"{ticker}_sma_profit"] = sma_profit
        results[f"{ticker}_sma_returns"] = sma_returns

    except FileNotFoundError:
        print(f"Missing data file for {ticker}")
        continue

# save the results dictionary to a file called results.json
saveResults(results)
print("\nSimulation complete! Results saved to results.json")
```
